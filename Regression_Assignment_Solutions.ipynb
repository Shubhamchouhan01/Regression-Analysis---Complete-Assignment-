## Part 1: Simple Linear Regression

### Question 1: What is Simple Linear Regression?

**Answer:**

Simple Linear Regression is a statistical method to model the linear relationship between a single independent variable (X) and a dependent variable (Y). It assumes a straight-line relationship between the two variables.

**Equation:** Y = mX + c

Where:
- Y = Dependent variable (output/target)
- X = Independent variable (input/predictor)
- m = Slope of the line
- c = Y-intercept (constant term)

### Question 2: What are the key assumptions of Simple Linear Regression?

**Answer:**

1. **Linearity**: The relationship between X and Y is linear.
2. **Independence**: Observations are independent of each other.
3. **Homoscedasticity**: The variance of residuals is constant across all levels of X.
4. **Normality**: Residuals are normally distributed.
5. **No Multicollinearity**: Only applicable in multiple regression, but SLR has only one predictor.
6. **No Autocorrelation**: Residuals are not correlated with each other.

### Question 3: What does the coefficient m represent in the equation Y=mX+c?

**Answer:**

The coefficient **m** is the **slope** of the regression line.

**Interpretation:**
- It represents the change in Y for every one-unit change in X.
- If m = 2, then for every 1-unit increase in X, Y increases by 2 units.
- If m = -0.5, then for every 1-unit increase in X, Y decreases by 0.5 units.
- m = 0 means no relationship between X and Y.

### Question 4: What does the intercept c represent in the equation Y=mX+c?

**Answer:**

The intercept **c** is the **Y-intercept** or the constant term.

**Interpretation:**
- It is the value of Y when X = 0.
- It represents the baseline value of the dependent variable.
- It shifts the regression line up or down on the Y-axis.
- In practical terms, it's the expected value of Y when the predictor has no contribution.

### Question 5: How do we calculate the slope m in Simple Linear Regression?

**Answer:**

The slope m is calculated using the formula:

$$m = \frac{\sum_{i=1}^{n}(X_i - \bar{X})(Y_i - \bar{Y})}{\sum_{i=1}^{n}(X_i - \bar{X})^2}$$

Or equivalently:

$$m = \frac{n\sum X_i Y_i - \sum X_i \sum Y_i}{n\sum X_i^2 - (\sum X_i)^2}$$

Where:
- n = number of data points
- X̄ = mean of X values
- Ȳ = mean of Y values

**Once m is calculated, c is computed as:**
$$c = \bar{Y} - m\bar{X}$$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import warnings
warnings.filterwarnings('ignore')

# Example: Calculate slope manually
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 4, 5, 4, 6])

n = len(X)
X_mean = np.mean(X)
Y_mean = np.mean(Y)

# Calculate slope
numerator = np.sum((X - X_mean) * (Y - Y_mean))
denominator = np.sum((X - X_mean) ** 2)
m = numerator / denominator

# Calculate intercept
c = Y_mean - m * X_mean

print(f"Slope (m): {m:.4f}")
print(f"Intercept (c): {c:.4f}")
print(f"Regression Equation: Y = {m:.4f}X + {c:.4f}")
```

### Question 6: What is the purpose of the least squares method in Simple Linear Regression?

**Answer:**

The **Least Squares Method** is a technique to find the best-fit line that minimizes the sum of squared residuals.

**Purpose:**
1. **Minimize Error**: It finds the regression line that best explains the relationship between X and Y by minimizing the squared differences between observed values and predicted values.
2. **Optimal Fit**: It ensures the regression line passes through the center of the data (at the mean point).
3. **Mathematical Optimization**: It provides a unique and optimal solution for the coefficients m and c.

**Formula:**
$$\text{Minimize} \sum_{i=1}^{n}(Y_i - \hat{Y}_i)^2 = \sum_{i=1}^{n}(Y_i - (mX_i + c))^2$$

Where:
- Yi = Actual value
- Ŷi = Predicted value
- (Yi - Ŷi) = Residual

### Question 7: How is the coefficient of determination (R²) interpreted in Simple Linear Regression?

**Answer:**

**R² (Coefficient of Determination)** measures the proportion of variance in the dependent variable that is explained by the independent variable.

**Formula:**
$$R^2 = 1 - \frac{\sum_{i=1}^{n}(Y_i - \hat{Y}_i)^2}{\sum_{i=1}^{n}(Y_i - \bar{Y})^2} = \frac{\text{Explained Variance}}{\text{Total Variance}}$$

**Interpretation:**
- **Range**: 0 to 1 (0% to 100%)
- **R² = 1**: Perfect fit - the model explains 100% of the variance
- **R² = 0.8**: The model explains 80% of the variance in Y
- **R² = 0**: The model explains 0% of the variance
- **Higher R² is better**, but doesn't guarantee a good model

**Limitations:**
- Increases with more variables (even if they're not useful)
- Doesn't indicate causation
- Affected by outliers

```python
# Calculate R² for the example data
Y_pred = m * X + c

# Sum of squared residuals
SS_res = np.sum((Y - Y_pred) ** 2)

# Total sum of squares
SS_tot = np.sum((Y - Y_mean) ** 2)

# Calculate R²
R_squared = 1 - (SS_res / SS_tot)

print(f"R² Score: {R_squared:.4f}")
print(f"\nInterpretation: The model explains {R_squared*100:.2f}% of the variance in Y")

# Visualization
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Actual Data', s=100)
plt.plot(X, Y_pred, color='red', label='Regression Line', linewidth=2)

# Show residuals
for i in range(len(X)):
    plt.plot([X[i], X[i]], [Y[i], Y_pred[i]], 'g--', alpha=0.5)

plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)
plt.title(f'Simple Linear Regression (R² = {R_squared:.4f})', fontsize=14)
plt.legend(fontsize=10)
plt.grid(alpha=0.3)
plt.show()
```

## Part 2: Multiple Linear Regression

### Question 8: What is Multiple Linear Regression?

**Answer:**

Multiple Linear Regression extends Simple Linear Regression to model the relationship between a dependent variable (Y) and **multiple independent variables** (X1, X2, X3, ..., Xn).

**Equation:**
$$Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \beta_3 X_3 + ... + \beta_n X_n + \epsilon$$

Where:
- Y = Dependent variable
- X1, X2, ..., Xn = Independent variables
- β0 = Intercept
- β1, β2, ..., βn = Coefficients/Slopes
- ε = Error term

**Key Difference from SLR:**
- Uses multiple predictors instead of just one
- More complex but often more accurate
- Better captures real-world relationships

### Question 9: What is the main difference between Simple and Multiple Linear Regression?

**Answer:**

| Aspect | Simple Linear Regression | Multiple Linear Regression |
|--------|--------------------------|---------------------------|
| **Number of Predictors** | 1 independent variable | 2+ independent variables |
| **Equation** | Y = mX + c | Y = β0 + β1X1 + β2X2 + ... |
| **Complexity** | Simple, easy to interpret | More complex |
| **Visualization** | 2D plot (line) | Multi-dimensional |
| **Flexibility** | Limited | More flexible |
| **Accuracy** | Lower in complex scenarios | Generally higher |
| **Interpretation** | One slope value | Multiple coefficients |
| **Overfitting Risk** | Low | Higher with many variables |

### Question 10: What are the key assumptions of Multiple Linear Regression?

**Answer:**

1. **Linearity**: The relationship between Y and each Xi is linear.
2. **Independence**: Observations are independent of each other.
3. **Homoscedasticity**: Constant variance of residuals across all values.
4. **Normality**: Residuals follow a normal distribution.
5. **No Multicollinearity**: Independent variables are not highly correlated with each other.
6. **No Autocorrelation**: Residuals are not correlated with each other (especially for time-series data).
7. **No Outliers**: Extreme values don't unduly influence the model.
8. **Sufficient Sample Size**: Enough observations relative to number of variables.

### Question 11: What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?

**Answer:**

**Heteroscedasticity** is a violation of the homoscedasticity assumption where the variance of residuals is NOT constant across different levels of the independent variable(s).

**Effects on MLR Model:**
1. **Biased Standard Errors**: Standard errors of coefficients are underestimated or overestimated.
2. **Invalid Confidence Intervals**: Confidence intervals become unreliable.
3. **Inefficient Estimates**: OLS estimates are no longer the best linear unbiased estimators (BLUE).
4. **Incorrect P-values**: Hypothesis tests become unreliable.
5. **Unreliable Predictions**: Model predictions may be less reliable.

**Detection Methods:**
- Visual inspection of residual plots
- Breusch-Pagan test
- White's test

**Solutions:**
- Weighted Least Squares (WLS)
- Variable transformation
- Robust standard errors

```python
# Example: Multiple Linear Regression with visualization
np.random.seed(42)

# Create sample data
n_samples = 100
X1 = np.random.randn(n_samples)
X2 = np.random.randn(n_samples)
X3 = np.random.randn(n_samples)

# Create Y with heteroscedasticity
Y = 5 + 3*X1 + 2*X2 - 1*X3 + np.random.randn(n_samples) * (0.5 + np.abs(X1))

# Build MLR model
X_data = np.column_stack([X1, X2, X3])
model = LinearRegression()
model.fit(X_data, Y)
Y_pred = model.predict(X_data)

# Calculate residuals
residuals = Y - Y_pred

# Visualization: Residual plot (shows heteroscedasticity)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Residuals vs Fitted values
axes[0].scatter(Y_pred, residuals, alpha=0.6)
axes[0].axhline(y=0, color='r', linestyle='--', linewidth=2)
axes[0].set_xlabel('Fitted Values', fontsize=12)
axes[0].set_ylabel('Residuals', fontsize=12)
axes[0].set_title('Residual Plot (Shows Heteroscedasticity)', fontsize=12)
axes[0].grid(alpha=0.3)

# Plot 2: Q-Q plot for normality
from scipy import stats
stats.probplot(residuals, dist="norm", plot=axes[1])
axes[1].set_title('Q-Q Plot (Normality Check)', fontsize=12)
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

print(f"Intercept: {model.intercept_:.4f}")
print(f"Coefficients: {model.coef_}")
print(f"R² Score: {r2_score(Y, Y_pred):.4f}")
```

### Question 12: How can you improve a Multiple Linear Regression model with high multicollinearity?

**Answer:**

**Multicollinearity** occurs when independent variables are highly correlated with each other.

**Improvement Techniques:**

1. **Remove Correlated Variables**
   - Identify and remove one of the correlated variables
   - Use correlation matrix to find problematic pairs

2. **Combine Variables**
   - Create new variables from correlated ones
   - Use Principal Component Analysis (PCA)

3. **Ridge Regression**
   - Adds a penalty term to regression: Cost = RSS + λ(sum of squared coefficients)
   - Shrinks coefficients of correlated variables

4. **Lasso Regression**
   - Uses L1 penalty: Cost = RSS + λ(sum of absolute coefficients)
   - Can set some coefficients to exactly zero

5. **Elastic Net**
   - Combines Ridge and Lasso
   - Better for high-dimensional data

6. **VIF (Variance Inflation Factor)**
   - Calculate VIF for each variable
   - Remove variables with VIF > 5 or 10

7. **Feature Selection**
   - Use forward/backward selection
   - Select features with highest importance

```python
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.preprocessing import StandardScaler

# Example: Comparing different regression techniques with multicollinearity
np.random.seed(42)

# Create data with multicollinearity
n_samples = 100
X1 = np.random.randn(n_samples)
X2 = X1 + np.random.randn(n_samples) * 0.1  # Highly correlated with X1
X3 = np.random.randn(n_samples)

Y = 5 + 3*X1 + 2*X2 + X3 + np.random.randn(n_samples) * 2

X_data = np.column_stack([X1, X2, X3])

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_data)

# Calculate correlation matrix to show multicollinearity
df = pd.DataFrame(X_data, columns=['X1', 'X2', 'X3'])
print("Correlation Matrix:")
print(df.corr())
print("\nNote: X1 and X2 are highly correlated (0.9+)")

# Fit different models
# 1. Ordinary Linear Regression
lr_model = LinearRegression()
lr_model.fit(X_scaled, Y)

# 2. Ridge Regression (L2 penalty)
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_scaled, Y)

# 3. Lasso Regression (L1 penalty)
lasso_model = Lasso(alpha=0.1)
lasso_model.fit(X_scaled, Y)

# 4. Elastic Net
elasticnet_model = ElasticNet(alpha=0.1, l1_ratio=0.5)
elasticnet_model.fit(X_scaled, Y)

# Compare coefficients
print("\n" + "="*60)
print("Coefficient Comparison:")
print("="*60)
print(f"OLS Coefficients:       {lr_model.coef_}")
print(f"Ridge Coefficients:     {ridge_model.coef_}")
print(f"Lasso Coefficients:     {lasso_model.coef_}")
print(f"ElasticNet Coefficients:{elasticnet_model.coef_}")

# Visualization
fig, ax = plt.subplots(figsize=(10, 6))
x_pos = np.arange(3)
width = 0.2

ax.bar(x_pos - 1.5*width, lr_model.coef_, width, label='OLS')
ax.bar(x_pos - 0.5*width, ridge_model.coef_, width, label='Ridge')
ax.bar(x_pos + 0.5*width, lasso_model.coef_, width, label='Lasso')
ax.bar(x_pos + 1.5*width, elasticnet_model.coef_, width, label='ElasticNet')

ax.set_xlabel('Features', fontsize=12)
ax.set_ylabel('Coefficient Value', fontsize=12)
ax.set_title('Coefficient Comparison: Different Regression Methods', fontsize=14)
ax.set_xticks(x_pos)
ax.set_xticklabels(['X1', 'X2', 'X3'])
ax.legend()
ax.grid(alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

### Question 14: What is the role of interaction terms in Multiple Linear Regression?

**Answer:**

**Interaction Terms** represent the combined effect of two or more variables on the dependent variable.

**Definition:**
An interaction term is the product of two or more independent variables.

**Equation with Interaction:**
$$Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \beta_3 (X_1 \times X_2) + \epsilon$$

**Key Roles:**
1. **Capture Non-additive Effects**: Allows one variable's effect to depend on another variable
2. **Improve Model Fit**: Can increase R² if relevant interactions exist
3. **More Realistic Models**: Real-world relationships are often interactive
4. **Better Predictions**: More accurate for complex scenarios

**Example:**
- House Price = β0 + β1(Size) + β2(Location) + β3(Size × Location) + ε
- Size effect on price depends on location (urban vs rural)

**Interpretation of Coefficients:**
- β3 represents how the effect of X1 changes with X2
- If β3 > 0: interaction is synergistic
- If β3 < 0: interaction is antagonistic

**Cautions:**
- Increases model complexity
- Can introduce multicollinearity
- Harder to interpret
- Use only when theoretically justified

```python
# Example: Interaction terms in regression
np.random.seed(42)

# Create sample data
X1 = np.random.randn(100)  # Feature 1
X2 = np.random.randn(100)  # Feature 2

# Y with interaction effect
Y = 10 + 3*X1 + 2*X2 + 4*(X1*X2) + np.random.randn(100) * 2  # Interaction term

# Model WITHOUT interaction
X_without_interaction = np.column_stack([X1, X2])
model_without = LinearRegression()
model_without.fit(X_without_interaction, Y)
Y_pred_without = model_without.predict(X_without_interaction)
r2_without = r2_score(Y, Y_pred_without)

# Model WITH interaction
X1_X2_interaction = X1 * X2
X_with_interaction = np.column_stack([X1, X2, X1_X2_interaction])
model_with = LinearRegression()
model_with.fit(X_with_interaction, Y)
Y_pred_with = model_with.predict(X_with_interaction)
r2_with = r2_score(Y, Y_pred_with)

print("Comparison: Without vs With Interaction Term")
print("="*60)
print(f"\nModel WITHOUT Interaction:")
print(f"  Coefficients: β0={model_without.intercept_:.4f}, β1={model_without.coef_[0]:.4f}, β2={model_without.coef_[1]:.4f}")
print(f"  R² Score: {r2_without:.4f}")

print(f"\nModel WITH Interaction:")
print(f"  Coefficients: β0={model_with.intercept_:.4f}, β1={model_with.coef_[0]:.4f}, β2={model_with.coef_[1]:.4f}, β3(interaction)={model_with.coef_[2]:.4f}")
print(f"  R² Score: {r2_with:.4f}")

print(f"\nImprovement: R² increased by {(r2_with - r2_without):.4f}")
print(f"\nInterpretation:")
print(f"  The interaction coefficient ({model_with.coef_[2]:.4f}) suggests that")
print(f"  the effect of X1 on Y depends on the value of X2 and vice versa.")
```

### Question 15: How can the interpretation of intercept differ between Simple and Multiple Linear Regression?

**Answer:**

| Aspect | Simple Linear Regression | Multiple Linear Regression |
|--------|--------------------------|---------------------------|
| **Definition** | Y value when X = 0 | Y value when ALL Xi = 0 |
| **Formula** | c in Y = mX + c | β0 in Y = β0 + β1X1 + β2X2 + ... |
| **Interpretation** | Baseline value of Y (single predictor) | Baseline value of Y (all predictors zero) |
| **Practical Meaning** | Often has direct interpretation | May or may not be meaningful |
| **Example** | Salary at 0 years experience | Salary at 0 years exp, 0 education, 0 skills... |
| **Reality Check** | Sometimes unrealistic | Usually unrealistic (all variables rarely = 0) |

**Key Differences:**

1. **Context Dependency**
   - SLR: Can often be interpreted directly
   - MLR: Interpretation depends on whether all variables can be zero simultaneously

2. **Meaningfulness**
   - SLR: Usually has practical meaning
   - MLR: Often lacks practical meaning (all predictors rarely = 0)

3. **Use in Predictions**
   - Both: Essential for making predictions
   - But MLR intercept rarely represents a realistic scenario

### Question 16: What is the significance of the slope in regression analysis, and how does it affect predictions?

**Answer:**

**Slope Significance:**

The **slope (m or β)** represents the rate of change of the dependent variable with respect to the independent variable.

**Key Aspects:**

1. **Magnitude**
   - Larger slope = stronger relationship
   - Small slope = weaker relationship
   - Slope = 0 means no relationship

2. **Direction**
   - Positive slope: Y increases as X increases
   - Negative slope: Y decreases as X increases

3. **Impact on Predictions**
   - Changes in slope alter predicted values significantly
   - Steep slope → small X changes cause large Y changes
   - Flat slope → X changes have minimal impact on Y

4. **Statistical Significance**
   - Test if slope is significantly different from zero
   - Use t-test and p-value
   - Significant if p-value < 0.05

5. **Standard Error**
   - Lower SE = more reliable slope estimate
   - Higher SE = less reliable estimate

```python
# Example: Impact of slope on predictions
from scipy import stats

# Create data for analysis
X = np.array([1, 2, 3, 4, 5, 6, 7, 8])
Y = np.array([2.1, 4.2, 6.1, 7.8, 10.2, 11.9, 14.1, 15.9])

# Fit regression model
model = LinearRegression()
model.fit(X.reshape(-1, 1), Y)
slope = model.coef_[0]
intercept = model.intercept_

# Calculate statistics
Y_pred = model.predict(X.reshape(-1, 1))
residuals = Y - Y_pred
se_residuals = np.std(residuals)

# Calculate standard error of slope
ss_x = np.sum((X - np.mean(X))**2)
se_slope = se_residuals / np.sqrt(ss_x)

# t-statistic and p-value
t_stat = slope / se_slope
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), len(X) - 2))

# R-squared
r_squared = r2_score(Y, Y_pred)

print("REGRESSION ANALYSIS RESULTS")
print("="*70)
print(f"\nEquation: Y = {intercept:.4f} + {slope:.4f}X")
print(f"\nSlope Analysis:")
print(f"  Slope (β1): {slope:.4f}")
print(f"  Interpretation: For every 1-unit increase in X, Y increases by {slope:.4f} units")
print(f"  Standard Error of Slope: {se_slope:.4f}")
print(f"  t-statistic: {t_stat:.4f}")
print(f"  p-value: {p_value:.6f}")

if p_value < 0.05:
    print(f"  ✓ Slope is STATISTICALLY SIGNIFICANT (p < 0.05)")
else:
    print(f"  ✗ Slope is NOT statistically significant (p >= 0.05)")

print(f"\nR² Score: {r_squared:.4f}")
print(f"Model Quality: {r_squared*100:.2f}% of variance explained")

# Prediction examples
print(f"\nPrediction Examples:")
print(f"  When X = 0: Y = {intercept:.4f}")
print(f"  When X = 5: Y = {intercept + slope*5:.4f}")
print(f"  When X = 10: Y = {intercept + slope*10:.4f}")

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Regression line and data
X_range = np.linspace(0, 10, 100)
Y_range = intercept + slope * X_range

axes[0].scatter(X, Y, color='blue', s=100, label='Actual Data')
axes[0].plot(X_range, Y_range, color='red', linewidth=2, label='Regression Line')
axes[0].set_xlabel('X', fontsize=12)
axes[0].set_ylabel('Y', fontsize=12)
axes[0].set_title(f'Regression Line (Slope = {slope:.4f})', fontsize=12)
axes[0].legend(fontsize=10)
axes[0].grid(alpha=0.3)

# Plot 2: Residuals
axes[1].scatter(Y_pred, residuals, color='green', s=100)
axes[1].axhline(y=0, color='red', linestyle='--', linewidth=2)
axes[1].set_xlabel('Fitted Values', fontsize=12)
axes[1].set_ylabel('Residuals', fontsize=12)
axes[1].set_title('Residual Plot', fontsize=12)
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

### Question 18: What are the limitations of using R² as a sole measure of model performance?

**Answer:**

**Limitations of R² as Sole Performance Measure:**

1. **Increases with More Variables**
   - R² never decreases when adding variables
   - Even irrelevant variables increase R²
   - Solution: Use Adjusted R²

2. **Doesn't Indicate Causation**
   - High R² doesn't mean X causes Y
   - Variables can be correlated without causal relationship
   - Still need domain knowledge

3. **Doesn't Guarantee Good Predictions**
   - High R² on training data can lead to overfitting
   - Poor performance on new data
   - Test on holdout data instead

4. **Sensitive to Outliers**
   - Extreme values can artificially inflate or deflate R²
   - May misrepresent actual fit quality

5. **Doesn't Assess Assumption Violations**
   - High R² doesn't mean assumptions are met
   - Must check linearity, homoscedasticity, normality separately

6. **Scale-Dependent**
   - R² value depends on outcome variable scale
   - Not comparable across different outcome variables

7. **Doesn't Check Significance**
   - High R² doesn't mean variables are statistically significant
   - Must check p-values and confidence intervals

**Better Alternative Metrics:**
- **Adjusted R²**: Penalizes adding variables
- **AIC/BIC**: Balanced model fit and complexity
- **RMSE**: Average prediction error
- **MAE**: Absolute average error
- **Cross-validation score**: Performance on test data

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Example: Why R² alone is not sufficient
np.random.seed(42)

n = 100
X_base = np.random.randn(n)
Y = 5 + 3*X_base + np.random.randn(n) * 2

# Add random features (no real relationship with Y)
X_random1 = np.random.randn(n)
X_random2 = np.random.randn(n)
X_random3 = np.random.randn(n)

# Model 1: Only meaningful variable
X1 = X_base.reshape(-1, 1)
model1 = LinearRegression()
model1.fit(X1, Y)
y_pred1 = model1.predict(X1)
r2_1 = r2_score(Y, y_pred1)
adj_r2_1 = 1 - (1 - r2_1) * (n - 1) / (n - 1 - 1)

# Model 2: Meaningful + random variables
X2 = np.column_stack([X_base, X_random1, X_random2, X_random3])
model2 = LinearRegression()
model2.fit(X2, Y)
y_pred2 = model2.predict(X2)
r2_2 = r2_score(Y, y_pred2)
adj_r2_2 = 1 - (1 - r2_2) * (n - 1) / (n - 1 - 4)

# Performance metrics
rmse1 = np.sqrt(mean_squared_error(Y, y_pred1))
rmse2 = np.sqrt(mean_squared_error(Y, y_pred2))
mae1 = mean_absolute_error(Y, y_pred1)
mae2 = mean_absolute_error(Y, y_pred2)

print("COMPARISON: Simple Model vs Complex Model with Irrelevant Variables")
print("="*80)

print(f"\nMODEL 1: Only meaningful variable (X_base)")
print("-"*80)
print(f"  R² Score:        {r2_1:.6f}")
print(f"  Adjusted R²:     {adj_r2_1:.6f}")
print(f"  RMSE:            {rmse1:.6f}")
print(f"  MAE:             {mae1:.6f}")
print(f"  Number of features: 1")

print(f"\nMODEL 2: Meaningful + 3 random variables")
print("-"*80)
print(f"  R² Score:        {r2_2:.6f}")
print(f"  Adjusted R²:     {adj_r2_2:.6f}")
print(f"  RMSE:            {rmse2:.6f}")
print(f"  MAE:             {mae2:.6f}")
print(f"  Number of features: 4")

print(f"\nKEY OBSERVATIONS:")
print("="*80)
print(f"  ✓ R² increased from {r2_1:.6f} to {r2_2:.6f} by adding random variables")
print(f"  ✗ Adjusted R² DECREASED from {adj_r2_1:.6f} to {adj_r2_2:.6f}")
print(f"  ✗ RMSE slightly increased (overfitting): {rmse1:.6f} → {rmse2:.6f}")
print(f"\nCONCLUSION:")
print(f"  Using R² alone would lead to choosing the worse model!")
print(f"  Adjusted R² correctly penalizes adding irrelevant variables.")
print(f"  Always use multiple metrics for model evaluation.")
```

### Question 19: How would you interpret a large standard error for a regression coefficient?

**Answer:**

**Large Standard Error (SE) Interpretation:**

**Definition:**
Standard Error measures the variability/uncertainty in the estimate of a regression coefficient.

**What Large SE Indicates:**

1. **High Uncertainty**
   - The estimated coefficient is not reliable
   - The true value could differ significantly from the estimate
   - Wide confidence interval around the estimate

2. **Less Precise Estimate**
   - More variability in the coefficient estimate
   - Less confident about the effect size
   - Confidence interval is wider

3. **Lower Statistical Significance**
   - t-statistic = Coefficient / SE
   - Larger SE → smaller t-statistic
   - Larger p-value (less significant)

**Causes of Large Standard Error:**

1. **Multicollinearity**: Predictors are highly correlated
2. **High Residual Variance**: Large unexplained variability in Y
3. **Small Sample Size**: Few observations to estimate the effect
4. **Small Variance in Predictor**: Little variation in X
5. **Weak Signal**: Weak relationship between X and Y

**Practical Implications:**
- **Not Statistically Significant**: Large SE → large p-value
- **Unreliable for Predictions**: Effect could be much smaller or larger
- **Needs Caution**: Cannot confidently claim the relationship exists

### Question 20: How can heteroscedasticity be identified in residual plots, and why is it important to address it?

```python
# Example: Detecting heteroscedasticity
np.random.seed(42)

# Create data with heteroscedasticity
n = 200
X = np.random.uniform(0, 10, n)
# Variance increases with X
Y = 5 + 2*X + np.random.normal(0, 1 + X, n)  # Variance depends on X

# Fit model
X_reshaped = X.reshape(-1, 1)
model = LinearRegression()
model.fit(X_reshaped, Y)
Y_pred = model.predict(X_reshaped)
residuals = Y - Y_pred

# Create diagnostic plots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Scatter with regression line
axes[0, 0].scatter(X, Y, alpha=0.5)
axes[0, 0].plot(X, Y_pred, color='red', linewidth=2)
axes[0, 0].set_xlabel('X')
axes[0, 0].set_ylabel('Y')
axes[0, 0].set_title('Scatter Plot with Regression Line')
axes[0, 0].grid(alpha=0.3)

# Plot 2: Residuals vs Fitted Values (SHOWS HETEROSCEDASTICITY)
axes[0, 1].scatter(Y_pred, residuals, alpha=0.6, color='red')
axes[0, 1].axhline(y=0, color='black', linestyle='--', linewidth=2)
axes[0, 1].set_xlabel('Fitted Values')
axes[0, 1].set_ylabel('Residuals')
axes[0, 1].set_title('Residual Plot (❌ Shows Heteroscedasticity - Funnel Shape)')
axes[0, 1].grid(alpha=0.3)

# Plot 3: Histogram of residuals
axes[1, 0].hist(residuals, bins=30, edgecolor='black', alpha=0.7, color='green')
axes[1, 0].set_xlabel('Residuals')
axes[1, 0].set_ylabel('Frequency')
axes[1, 0].set_title('Distribution of Residuals')
axes[1, 0].grid(alpha=0.3, axis='y')

# Plot 4: Q-Q Plot
stats.probplot(residuals, dist="norm", plot=axes[1, 1])
axes[1, 1].set_title('Q-Q Plot (Normality of Residuals)')
axes[1, 1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Breusch-Pagan Test
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.tools.tools import add_constant

X_with_const = add_constant(X)
bp_test = het_breuschpagan(residuals, X_with_const)

print("\nHETEROSCEDASTICITY TESTS")
print("="*70)
print(f"\nBreusch-Pagan Test:")
print(f"  Test Statistic: {bp_test[0]:.6f}")
print(f"  p-value: {bp_test[1]:.6f}")
if bp_test[1] < 0.05:
    print(f"  ✗ Result: HETEROSCEDASTICITY DETECTED (p < 0.05)")
else:
    print(f"  ✓ Result: No heteroscedasticity (p >= 0.05)")

print(f"\nInterpretation:")
print(f"  - Null Hypothesis: Homoscedasticity (equal variance)")
print(f"  - Alternative: Heteroscedasticity (unequal variance)")
print(f"  - p-value < 0.05: Reject null hypothesis → Heteroscedasticity present")
```

### Question 21: What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?

```python
# Example: Demonstrating overfitting with R² vs Adjusted R²
np.random.seed(42)

n_samples = 100
X_base = np.random.randn(n_samples)
Y = 5 + 3*X_base + np.random.randn(n_samples) * 2

# Generate many random features (noise)
n_features_max = 15
X_random = np.random.randn(n_samples, n_features_max)

# Build models with increasing number of features
r2_scores = []
adj_r2_scores = []
feature_counts = []

for p in range(1, n_features_max + 1):
    X = np.column_stack([X_base] + [X_random[:, i] for i in range(p-1)])
    
    model = LinearRegression()
    model.fit(X, Y)
    y_pred = model.predict(X)
    
    r2 = r2_score(Y, y_pred)
    adj_r2 = 1 - (1 - r2) * (n_samples - 1) / (n_samples - p - 1)
    
    r2_scores.append(r2)
    adj_r2_scores.append(adj_r2)
    feature_counts.append(p)

# Visualization
plt.figure(figsize=(12, 6))
plt.plot(feature_counts, r2_scores, marker='o', linewidth=2, label='R² (always increases)', color='red')
plt.plot(feature_counts, adj_r2_scores, marker='s', linewidth=2, label='Adjusted R² (penalizes extra variables)', color='green')
plt.axvline(x=1, color='blue', linestyle='--', alpha=0.5, label='Optimal (1 meaningful variable)')
plt.xlabel('Number of Features', fontsize=12)
plt.ylabel('R² Score', fontsize=12)
plt.title('R² vs Adjusted R² (Overfitting Example)', fontsize=14)
plt.legend(fontsize=11)
plt.grid(alpha=0.3)
plt.xticks(feature_counts)
plt.show()

# Find optimal number of features
optimal_p = feature_counts[np.argmax(adj_r2_scores)]

print("\nOVERFITTING ANALYSIS")
print("="*70)
print(f"\nOptimal number of features (max Adjusted R²): {optimal_p}")
print(f"  R² at optimal: {r2_scores[optimal_p-1]:.6f}")
print(f"  Adjusted R² at optimal: {adj_r2_scores[optimal_p-1]:.6f}")

print(f"\nWith maximum features ({n_features_max}):")
print(f"  R²: {r2_scores[-1]:.6f}")
print(f"  Adjusted R²: {adj_r2_scores[-1]:.6f}")

print(f"\nDifferences:")
print(f"  R² increased by: {(r2_scores[-1] - r2_scores[optimal_p-1]):.6f}")
print(f"  Adjusted R² decreased by: {(adj_r2_scores[optimal_p-1] - adj_r2_scores[-1]):.6f}")

print(f"\nConclusion:")
print(f"  ✗ R² increased, suggesting better model")
print(f"  ✓ Adjusted R² decreased, indicating overfitting")
print(f"  ✓ Should use model with {optimal_p} feature(s)")
```

### Question 22: Why is it important to scale variables in Multiple Linear Regression?

**Answer:**

**Why Scaling Variables is Important in MLR:**

1. **Fair Coefficient Comparison**
   - Unscaled: coefficients reflect different units
   - Scaled: coefficients are on same scale (standardized)
   - Can directly compare importance of variables

2. **Improves Numerical Stability**
   - Large differences in scale cause numerical errors
   - Algorithms more stable with scaled features
   - Reduces computational issues

3. **Faster Convergence**
   - Gradient descent (in optimization) converges faster
   - Fewer iterations needed
   - Better computational efficiency

4. **Reduces Multicollinearity Effects**
   - Scaling can reduce some multicollinearity issues
   - Easier to identify problematic variables

5. **Regularization Works Better**
   - Ridge/Lasso penalizes based on scale
   - Scaling ensures fair penalization
   - Prevents large-scale variables from dominating

6. **Meaningful Interpretation**
   - Scaled coefficients show relative importance
   - Easier to communicate results

**Scaling Methods:**
1. **Standardization (Z-score)**: X_scaled = (X - mean) / std
2. **Min-Max Scaling**: X_scaled = (X - min) / (max - min)
3. **Log Transformation**: For skewed data

**Note:** Scaling does NOT affect R² or model predictions for linear regression, but affects coefficient interpretation.

```python
# Example: Impact of scaling on coefficients and interpretation
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Create sample data with different scales
np.random.seed(42)
n = 100

# Features with different scales
Age = np.random.uniform(20, 70, n)  # 20-70 years
Income = np.random.uniform(30000, 150000, n)  # $30k-$150k
Experience = np.random.uniform(0, 40, n)  # 0-40 years

# Target variable
Salary = 20000 + 1000*Age + 0.01*Income + 2000*Experience + np.random.randn(n)*10000

X_unscaled = np.column_stack([Age, Income, Experience])

# Model 1: Unscaled data
model_unscaled = LinearRegression()
model_unscaled.fit(X_unscaled, Salary)

# Model 2: Standardized data
scaler_standard = StandardScaler()
X_scaled = scaler_standard.fit_transform(X_unscaled)
model_scaled = LinearRegression()
model_scaled.fit(X_scaled, Salary)

# Model 3: Min-Max scaled data
scaler_minmax = MinMaxScaler()
X_minmax = scaler_minmax.fit_transform(X_unscaled)
model_minmax = LinearRegression()
model_minmax.fit(X_minmax, Salary)

# Predictions (same for all models)
y_pred_unscaled = model_unscaled.predict(X_unscaled)
y_pred_scaled = model_scaled.predict(X_scaled)
y_pred_minmax = model_minmax.predict(X_minmax)

print("IMPACT OF SCALING ON COEFFICIENTS")
print("="*90)

print(f"\n1. UNSCALED DATA (Original Units)")
print("-"*90)
print(f"  Coefficients:")
print(f"    Age:        {model_unscaled.coef_[0]:>12.4f}  (salary change per year of age)")
print(f"    Income:     {model_unscaled.coef_[1]:>12.8f}  (salary change per dollar of income)")
print(f"    Experience:{model_unscaled.coef_[2]:>12.4f}  (salary change per year of experience)")
print(f"  Intercept:    {model_unscaled.intercept_:>12.4f}")
print(f"  R² Score:     {r2_score(Salary, y_pred_unscaled):>12.6f}")
print(f"\n  Problem: Can't compare coefficients due to different scales!")

print(f"\n2. STANDARDIZED DATA (Zero mean, Unit variance)")
print("-"*90)
print(f"  Coefficients (Standardized):")
print(f"    Age:        {model_scaled.coef_[0]:>12.4f}  (change in salary per 1 SD increase in Age)")
print(f"    Income:     {model_scaled.coef_[1]:>12.4f}  (change in salary per 1 SD increase in Income)")
print(f"    Experience:{model_scaled.coef_[2]:>12.4f}  (change in salary per 1 SD increase in Experience)")
print(f"  Intercept:    {model_scaled.intercept_:>12.4f}")
print(f"  R² Score:     {r2_score(Salary, y_pred_scaled):>12.6f}")
print(f"\n  ✓ Now comparable! Experience has largest effect (relative importance)")
print(f"  ✓ Same R² (scaling doesn't affect fit quality)")

print(f"\n3. MIN-MAX SCALED DATA (Range 0-1)")
print("-"*90)
print(f"  Coefficients (0-1 scale):")
print(f"    Age:        {model_minmax.coef_[0]:>12.4f}")
print(f"    Income:     {model_minmax.coef_[1]:>12.4f}")
print(f"    Experience:{model_minmax.coef_[2]:>12.4f}")
print(f"  Intercept:    {model_minmax.intercept_:>12.4f}")
print(f"  R² Score:     {r2_score(Salary, y_pred_minmax):>12.6f}")

print(f"\nKEY INSIGHTS:")
print("="*90)
print(f"  • R² is IDENTICAL for all models (scaling doesn't affect fit quality)")
print(f"  • Predictions are IDENTICAL (same relationships, different scale)")
print(f"  • Coefficients change, but relationships are preserved")
print(f"  • Standardized coefficients show relative importance of features")
print(f"  • Scaling is essential for fair comparison and regularization methods")
```

## Part 3: Polynomial Regression

### Question 23: What is polynomial regression?

**Answer:**

**Polynomial Regression** is a type of regression analysis where the relationship between the independent variable (X) and dependent variable (Y) is modeled as an nth-degree polynomial.

**Equation:**
$$Y = \beta_0 + \beta_1 X + \beta_2 X^2 + \beta_3 X^3 + ... + \beta_n X^n + \epsilon$$

**Key Characteristics:**
1. **Non-linear**: Can model curved relationships
2. **Still linear in parameters**: Can use linear regression methods
3. **Multiple features**: Creates polynomial features from single X
4. **Flexible**: Higher degree = more flexible fit

**Examples:**
- **Degree 1**: Y = β0 + β1X (Linear)
- **Degree 2**: Y = β0 + β1X + β2X² (Quadratic)
- **Degree 3**: Y = β0 + β1X + β2X² + β3X³ (Cubic)
- **Degree n**: n-th degree polynomial

**Advantages:**
- Captures curved relationships
- More flexible than linear regression
- Uses OLS for estimation (like linear regression)

**Disadvantages:**
- Can overfit with high degrees
- Harder to interpret coefficients
- Predictions become unstable outside data range

### Question 24: How does polynomial regression differ from linear regression?

**Answer:**

| Aspect | Linear Regression | Polynomial Regression |
|--------|-------------------|----------------------|
| **Relationship** | Straight line (Y = mX + c) | Curved line (nth degree) |
| **Flexibility** | Low | High |
| **Equation** | Y = β0 + β1X | Y = β0 + β1X + β2X² + ... |
| **Features** | 1 independent variable | Multiple polynomial features |
| **Fitting** | Simple, unique solution | More complex |
| **Overfitting Risk** | Low | High (especially high degree) |
| **Interpretation** | Simple | Complex |
| **When to Use** | Linear patterns | Curved/non-linear patterns |
| **Extrapolation** | Reasonable | Unreliable outside range |
| **Computational Cost** | Low | Moderate |
| **Assumption Violations** | Can violate linearity assumption | More flexible to non-linearity |

**Visual Difference:**
- Linear: Straight line through data
- Polynomial (degree 2): Parabola (U-shaped or inverted)
- Polynomial (degree 3+): More curves and bends

### Question 28: What are the limitations of polynomial regression?

**Answer:**

**Limitations of Polynomial Regression:**

1. **Overfitting**
   - High-degree polynomials fit noise, not signal
   - Excellent fit on training data, poor on test data
   - Risk increases dramatically with degree

2. **Extrapolation Problems**
   - Unreliable predictions outside data range
   - High-degree polynomials oscillate wildly
   - Cannot safely predict beyond observed data

3. **Interpretability**
   - Difficult to interpret coefficients
   - Especially hard with many variables and interactions
   - No clear "effect size" like linear regression

4. **Curse of Dimensionality**
   - Number of features grows exponentially
   - 2 variables, degree 3 = 10 features
   - 5 variables, degree 2 = 21 features
   - Computational complexity increases

5. **Data Requirements**
   - Need more observations with more features
   - Rule of thumb: n > 10p (n=observations, p=features)
   - Sparse data → unreliable estimates

6. **Multicollinearity**
   - Polynomial features are often correlated
   - Example: X and X² are highly correlated
   - Makes coefficient estimates unstable

7. **Loss of Simplicity**
   - Violates principle of parsimony
   - Complex model without proven necessity
   - Harder to communicate results

8. **Assumption Violations**
   - Still assumes linearity in parameters
   - Must check other OLS assumptions
   - High variance in residuals common

9. **Numerical Instability**
   - High-degree polynomials cause numerical problems
   - Matrix inversion becomes unstable
   - Results may be unreliable

10. **Model Selection Difficulty**
    - How to choose optimal degree?
    - Higher R² doesn't always mean better
    - Need cross-validation and test data

### Question 29: What methods can be used to evaluate model fit when selecting the degree of a polynomial?

```python
from sklearn.model_selection import cross_val_score, KFold

# Example: Model selection using multiple criteria
np.random.seed(42)

# Generate data
X = np.linspace(0, 10, 100).reshape(-1, 1)
Y = 5 + 2*X.ravel() - 0.15*X.ravel()**2 + np.random.randn(100)*2

# Evaluate different degrees
degrees = range(1, 11)
r2_train = []
adj_r2_train = []
cv_scores = []
aics = []
bics = []
rmses = []

from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score

for degree in degrees:
    # Create polynomial features
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)
    
    # Fit model
    model = LinearRegression()
    model.fit(X_poly, Y)
    
    # R² and Adjusted R²
    y_pred = model.predict(X_poly)
    r2 = r2_score(Y, y_pred)
    adj_r2 = 1 - (1 - r2) * (len(Y) - 1) / (len(Y) - degree - 1)
    
    # Cross-validation
    cv_score = cross_val_score(model, X_poly, Y, cv=5, scoring='r2').mean()
    
    # AIC and BIC
    residuals = Y - y_pred
    rss = np.sum(residuals**2)
    p = degree + 1  # number of parameters
    n = len(Y)
    
    aic = 2*p + n*np.log(rss/n)
    bic = p*np.log(n) + n*np.log(rss/n)
    rmse = np.sqrt(rss/n)
    
    # Store results
    r2_train.append(r2)
    adj_r2_train.append(adj_r2)
    cv_scores.append(cv_score)
    aics.append(aic)
    bics.append(bic)
    rmses.append(rmse)

# Visualization
fig, axes = plt.subplots(2, 3, figsize=(15, 8))

# Plot 1: R² and Adjusted R²
axes[0, 0].plot(degrees, r2_train, marker='o', label='R² (Training)', linewidth=2)
axes[0, 0].plot(degrees, adj_r2_train, marker='s', label='Adjusted R²', linewidth=2)
axes[0, 0].set_xlabel('Polynomial Degree')
axes[0, 0].set_ylabel('R² Score')
axes[0, 0].set_title('R² vs Adjusted R²')
axes[0, 0].legend()
axes[0, 0].grid(alpha=0.3)
axes[0, 0].set_xticks(degrees)

# Plot 2: Cross-validation score
axes[0, 1].plot(degrees, cv_scores, marker='o', color='green', linewidth=2)
axes[0, 1].set_xlabel('Polynomial Degree')
axes[0, 1].set_ylabel('CV Score (5-fold R²)')
axes[0, 1].set_title('Cross-Validation Score')
axes[0, 1].grid(alpha=0.3)
axes[0, 1].set_xticks(degrees)

# Plot 3: AIC and BIC
axes[0, 2].plot(degrees, aics, marker='o', label='AIC', linewidth=2)
axes[0, 2].plot(degrees, bics, marker='s', label='BIC', linewidth=2)
axes[0, 2].set_xlabel('Polynomial Degree')
axes[0, 2].set_ylabel('Information Criterion')
axes[0, 2].set_title('AIC and BIC (Lower is Better)')
axes[0, 2].legend()
axes[0, 2].grid(alpha=0.3)
axes[0, 2].set_xticks(degrees)

# Plot 4: RMSE
axes[1, 0].plot(degrees, rmses, marker='o', color='red', linewidth=2)
axes[1, 0].set_xlabel('Polynomial Degree')
axes[1, 0].set_ylabel('RMSE')
axes[1, 0].set_title('Root Mean Squared Error (Lower is Better)')
axes[1, 0].grid(alpha=0.3)
axes[1, 0].set_xticks(degrees)

# Plot 5: Summary comparison
axes[1, 1].axis('off')
summary_text = f"""OPTIMAL DEGREE SELECTION:


by Adjusted R²: Degree {np.argmax(adj_r2_train)+1}
by AIC: Degree {np.argmin(aics)+1}
by BIC: Degree {np.argmin(bics)+1}
by CV Score: Degree {np.argmax(cv_scores)+1}
by RMSE: Degree {np.argmin(rmses)+1}

RECOMMENDATION:
Degree 2 or 3 (True is degree 2)

Reasons:
• Multiple criteria agree
• Good generalization
• Interpretable
• Avoids overfitting"""
axes[1, 1].text(0.1, 0.5, summary_text, fontsize=11, verticalalignment='center',
                 bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.7))

# Plot 6: The fit comparison
X_range = np.linspace(0, 10, 200).reshape(-1, 1)
colors = plt.cm.viridis(np.linspace(0, 1, len(degrees)))

axes[1, 2].scatter(X, Y, color='black', s=50, label='Data', zorder=5)

for i, degree in enumerate([1, 2, 3, 5, 10]):
    poly = PolynomialFeatures(degree=degree)
    X_range_poly = poly.fit_transform(X_range)
    model = LinearRegression()
    X_poly = poly.fit_transform(X)
    model.fit(X_poly, Y)
    y_range_pred = model.predict(X_range_poly)
    axes[1, 2].plot(X_range, y_range_pred, label=f'Degree {degree}', alpha=0.7, linewidth=2)

axes[1, 2].set_xlabel('X')
axes[1, 2].set_ylabel('Y')
axes[1, 2].set_title('Polynomial Fits (Different Degrees)')
axes[1, 2].legend(fontsize=9)
axes[1, 2].grid(alpha=0.3)

plt.tight_layout()
plt.show()

print("\nMODEL SELECTION CRITERIA RESULTS:")
print("="*80)
for degree in degrees:
    idx = degree - 1
    print(f"\nDegree {degree}:")
    print(f"  R²:              {r2_train[idx]:>8.4f}")
    print(f"  Adjusted R²:     {adj_r2_train[idx]:>8.4f}")
    print(f"  CV Score (5-fold): {cv_scores[idx]:>8.4f}")
    print(f"  AIC:             {aics[idx]:>8.2f}")
    print(f"  BIC:             {bics[idx]:>8.2f}")
    print(f"  RMSE:            {rmses[idx]:>8.4f}")
```

### Question 31: How is polynomial regression implemented in Python?

```python
# COMPLETE POLYNOMIAL REGRESSION IMPLEMENTATION GUIDE

print("POLYNOMIAL REGRESSION IMPLEMENTATION IN PYTHON")
print("="*80)

# ============================================================================
# METHOD 1: Using NumPy (Manual Approach)
# ============================================================================
print("\nMETHOD 1: Manual Polynomial Features with NumPy")
print("-"*80)

X_data = np.array([1, 2, 3, 4, 5])
Y_data = np.array([2, 4, 5, 4, 6])

# Manually create polynomial features
X_poly_manual = np.column_stack([
    np.ones(len(X_data)),  # X^0 (constant)
    X_data,                 # X^1
    X_data**2              # X^2
])

model1 = LinearRegression(fit_intercept=False)  # We already included constant
model1.fit(X_poly_manual, Y_data)

print(f"Manual Polynomial (degree 2):")
print(f"  Coefficients: {model1.coef_}")
print(f"  Equation: Y = {model1.coef_[0]:.4f} + {model1.coef_[1]:.4f}X + {model1.coef_[2]:.4f}X²")

# ============================================================================
# METHOD 2: Using sklearn PolynomialFeatures (RECOMMENDED)
# ============================================================================
print("\n\nMETHOD 2: PolynomialFeatures (Recommended)")
print("-"*80)

from sklearn.preprocessing import PolynomialFeatures

# Reshape for sklearn (needs 2D)
X_reshaped = X_data.reshape(-1, 1)

# Create polynomial features
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X_reshaped)

# Fit model
model2 = LinearRegression()
model2.fit(X_poly, Y_data)

print(f"PolynomialFeatures (degree 2):")
print(f"  Intercept: {model2.intercept_:.6f}")
print(f"  Coefficients: {model2.coef_}")
print(f"  Feature names: {list(poly_features.get_feature_names_out(['X']))}")
print(f"  Equation: Y = {model2.intercept_:.4f} + {model2.coef_[0]:.4f}X + {model2.coef_[1]:.4f}X²")

# ============================================================================
# METHOD 3: Using Pipeline (Best Practice)
# ============================================================================
print("\n\nMETHOD 3: Pipeline (Best Practice - Prevents Data Leakage)")
print("-"*80)

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

pipeline = Pipeline([
    ('poly_features', PolynomialFeatures(degree=2)),
    ('scaler', StandardScaler()),
    ('linear_regression', LinearRegression())
])

pipeline.fit(X_reshaped, Y_data)
Y_pred_pipeline = pipeline.predict(X_reshaped)

print(f"Pipeline (Polynomial + Scaling + LR):")
print(f"  Predictions: {Y_pred_pipeline}")
print(f"  R² Score: {r2_score(Y_data, Y_pred_pipeline):.6f}")

# ============================================================================
# METHOD 4: Using numpy.polyfit() (Simple but Limited)
# ============================================================================
print("\n\nMETHOD 4: numpy.polyfit() (Simple for 1D)")
print("-"*80)

coefficients = np.polyfit(X_data, Y_data, deg=2)
poly_func = np.poly1d(coefficients)

print(f"numpy.polyfit (degree 2):")
print(f"  Coefficients: {coefficients}")
print(f"  Polynomial function: {poly_func}")
print(f"  Equation: Y = {coefficients[0]:.4f}X² + {coefficients[1]:.4f}X + {coefficients[2]:.4f}")
print(f"  Predictions: {poly_func(X_data)}")

# ============================================================================
# METHOD 5: Complete Example with Cross-Validation
# ============================================================================
print("\n\nMETHOD 5: Complete Example with Cross-Validation")
print("-"*80)

# Generate data
np.random.seed(42)
X_full = np.linspace(0, 10, 100).reshape(-1, 1)
Y_full = 2 + 3*X_full.ravel() - 0.2*X_full.ravel()**2 + np.random.randn(100)*2

# Create pipeline with polynomial features
polyPipeline = Pipeline([
    ('poly_features', PolynomialFeatures(degree=2)),
    ('linear_regression', LinearRegression())
])

# Perform cross-validation
from sklearn.model_selection import cross_val_score, train_test_split

cv_scores = cross_val_score(polyPipeline, X_full, Y_full, cv=5, scoring='r2')

print(f"5-Fold Cross-Validation Results:")
print(f"  CV Scores: {cv_scores}")
print(f"  Mean CV Score: {cv_scores.mean():.6f}")
print(f"  Std Dev: {cv_scores.std():.6f}")

# Train on full data for final model
polyPipeline.fit(X_full, Y_full)
Y_pred_full = polyPipeline.predict(X_full)

print(f"\n  Training R²: {r2_score(Y_full, Y_pred_full):.6f}")
print(f"  Training RMSE: {np.sqrt(np.mean((Y_full - Y_pred_full)**2)):.6f}")

# ============================================================================
# COMPARISON AND BEST PRACTICES
# ============================================================================
print("\n\nCOMPARISON OF METHODS")
print("="*80)

comparison_table = f"""
Method                  | Pros                        | Cons
-----------             | ---------                   | ----
1. NumPy (Manual)       | Full control, understands   | Tedious, error-prone
                        | math clearly                | 
-----------             | ---------                   | ----
2. PolynomialFeatures   | Automatic, flexible,        | Extra step needed
+ LinearRegression      | easy to use                 | for predictions
-----------             | ---------                   | ----
3. Pipeline             | Best practice, prevents     | Slightly more code
                        | data leakage                | 
-----------             | ---------                   | ----
4. numpy.polyfit()      | Simple, concise             | Limited flexibility
                        | Good for 1D                 | Hard to extend
-----------             | ---------                   | ----
5. Custom Functions     | Maximum flexibility         | Complex, slow
(scipy.optimize)        | for non-standard models     | 
"""

print(comparison_table)

print("\nBEST PRACTICES:")
print("-"*80)
print("1. USE PIPELINE: Prevents data leakage in cross-validation")
print("2. SCALE FEATURES: Important for regularization and numerical stability")
print("3. USE CROSS-VALIDATION: Evaluate on unseen data")
print("4. CHECK ASSUMPTIONS: Plot residuals, check normality")
print("5. AVOID HIGH DEGREES: Start with degree 2-3, increase only if needed")
print("6. USE REGULARIZATION: Ridge/Lasso for high-degree polynomials")
print("7. VISUALIZE: Always plot the data and fits")
```

## Summary and Key Takeaways

### Regression Analysis Overview

This notebook has covered three main types of regression:

1. **Simple Linear Regression**: Linear relationship between one predictor and outcome
2. **Multiple Linear Regression**: Linear relationships with multiple predictors
3. **Polynomial Regression**: Non-linear relationships using polynomial features

### Key Concepts to Remember

- **OLS (Ordinary Least Squares)**: Minimizes sum of squared residuals
- **R²**: Proportion of variance explained (0-1)
- **Adjusted R²**: Penalizes adding variables
- **Assumptions**: Linearity, independence, homoscedasticity, normality
- **Multicollinearity**: High correlation between predictors
- **Overfitting**: Model fits training data but doesn't generalize

### Best Practices

1. Always visualize data before modeling
2. Check assumptions through residual analysis
3. Use cross-validation for model evaluation
4. Scale features for regression methods
5. Compare multiple models and metrics
6. Avoid overfitting with regularization
7. Use Adjusted R² or AIC for model selection

### Tools and Libraries

- **scikit-learn**: LinearRegression, PolynomialFeatures, Pipeline
- **NumPy**: Mathematical operations and polynomial features
- **Pandas**: Data manipulation and exploration
- **Matplotlib/Seaborn**: Visualization
- **SciPy**: Statistical tests
